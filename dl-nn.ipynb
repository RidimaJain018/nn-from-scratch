```python
import numpy as np
```

```python
#Data
X = np.random.randn(4, 3)
y = np.array([0,1,1,1]).reshape(4,1)
```

```python
#intial weights and bias
weight = np.random.randn(3, 4)
bias = np.zeros((1, 4))
```

```python
#output weight and bias
weight_output = np.random.randn(4, 1)
bias_output = np.zeros((1, 1))
```

```python
#sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
lr = 0.01
```

```python
#looping 

for i in range(1000):
    #Hidden Layer:
    weighted_sum = X@weight + bias
    #ReLU function
    res = np.maximum(weighted_sum, 0)

    weighted_sum_output = res@weight_output + bias_output
    output = sigmoid(weighted_sum_output)
    
    # binary cross entropy
    loss = - np.mean((y) * np.log(output) + (1-y) * np.log(1-output))

    # BACKPROP - output layer
    dl_doutput = output - y
    dl_dweight_output = res.T @ dl_doutput
    dl_dbias_output = np.sum(dl_doutput, axis=0, keepdims=True)

    # BACKPROP - hidden layer
    dl_dres = dl_doutput @ weight_output.T
    dl_drelu = np.where(weighted_sum > 0, dl_dres, 0)
    dl_dweight = X.T @ dl_drelu
    dl_dbias = np.sum(dl_drelu, axis=0, keepdims=True)
    
    # WEIGHT UPDATE - gradient descent
    weight -= lr * dl_dweight
    bias -= lr * dl_dbias
    weight_output -= lr * dl_dweight_output
    bias_output -= lr * dl_dbias_output

    if i % 100 == 0: print(i, loss)
```
